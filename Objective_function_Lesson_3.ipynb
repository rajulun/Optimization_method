### Задание 1

Пусть у нас имеется m вариантов раскроя, т.е. известны величины $a_{ij} (i= \overline{1,m}, j=\overline{1,3})$, определяющие количество единиц j-x изделий при i-м способе раскроя единицы сырьевого материала и остаток трубы $a_i$.

Пусть $x_i$ - количество листов, раскраиваемых i-м вариантом $(i=\overline{1,m})$

Тогда количество изделий 1-го вида равно:

$$a_{11}x_1+...+a_{i1}x_i+...+a_{m1}x_m$$

Учитывая условия количества 

$$a_{11}x_1+...+a_{i1}x_i+...+a_{m1}x_m = 3000$$


Аналогичные равенства можно записать для остальных размеров.

Целевая функция же будет:
$$\sum_{i=1}^{m}x_ia_i \rightarrow min$$

$$a_{1j}x_1+...+a_{ij}x_i+...+a_{mj}x_m = k_j$$

где $k_j$ - необходимое количество раскроенных листов

Для начала решу именно эту задачу:

Все листы нестандартных размеров делят лист стандартного размера без остатков, кроме листа 1000×400 мм. Нам надо скомбинировать этот типоразмер с другим, чтобы не было остатков. У меня получилось разделить большой лист на 15 листов 1000×400 мм и 10 листов 500×600 мм. <br>
2000 / 15 = 133,3 листа нужно для нарезки 2000 листов 1000×400 мм <br>
133 * 15 = 1995  листов 1000×400 мм, остается 5 <br>
133 * 10 = 1330  листов 500×600 мм, остается 1670 <br>
1670 / 30 = 55,6 листа нужно для нарезки оставшихся листов 500×600 мм <br>
55 * 30 = 1650   листов 500×600 мм, остается еще 20 <br>
1500 / 6 = 250   листов нужно для нарезки 1500 листов 1500×1000 мм <br>
133 + 55 + 250 + 1 = 439 Всего нужно листов <br>
В остатке будет лист размером 1000×1000 мм



### Задание 2

5. Задача о курьерах

Пусть есть компания, занимающаяся доставкой Belivery Club.
Есть N зон, в которых есть спрос на курьеров в каждый час. Внутри дня в разные часы спрос различен. Каждой зоне i в час h необходимо необходимо $n_{hi}$ курьеров. Для каждой зоны есть штраф в случае, если не хватает 1 курьера - $s_i$

Необходимо распределить M курьеров таким образом, чтобы суммарный штраф был минимальный.



Пронумеруем курьеров:
пусть $x_{ih}^k$ = 1, если k-курьер работает в час h в i зоне

И их предпочтения:
пусть $y_{ih}^k$ = 1, если k-курьер предпочитает работать в час h в i зоне

Перебираем предпочтения
$$ f_{ih}^k = \begin{cases} 1, \space если \space y_{ih}^k = 1 \space \& \space x_{ih}^k = 1 \\
0, в\spaceпротивном\spaceслучае. \end {cases}$$

Тогда в одной зоне в час h работает 



$$\sum_{k=1}^{M} f_{ih}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} f_{ih}^k - n_{ih} >0 \\
s_i * (\sum_{k=1}^{M} x_{ih}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 



Штраф за день $$\sum_{h=1}^{24}f_{ih}$$

Штраф по всем зонам в течении дня:


$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}$$

Целевая функция:

$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih} \rightarrow min $$
